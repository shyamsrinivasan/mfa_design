```python
mutants = df_pairdist.columns

muts_none_all = []
for i in range(0, len(mutants)):
    mut1 = mutants[i]
    #print mut1 + ':',
    
    muts_none = []
    for j in range(i+1, len(mutants) - 1):
        mut2 = mutants[j]
        val = df_pairdist.loc[mut1, mut2]
        if val == 0:
            muts_none.append(mut2)
        
    #print ','.join(muts_none)
    muts_none_all += muts_none
    
muts_none_all = set(muts_none_all)
```

```python
[mut for mut in mutants if mut not in muts_none_all]
```
